```python
from IPython.display import clear_output

# 1) takes in an input
# 2) stores user input into a dictionary or list
# 3) the user can add or delete items
# 4) the user can see current shopping list
# 5) the program loops until the user 'quits'
# 6) upon quitting the program, prints out a receipt of the items with total and quantity

shopping_cart = {}

def go_shopping():
    total = 0
    name = input('What is your name? ').title()
    ask = input(f"Hi {name}! I'm hungry. I'm gonna go shopping for some food!\nWould you like to join me? y/n").lower().strip()
    clear_output()
    while ask not in ('y', 'n'):
        ask = input("Invalid response. Please enter y or n. ").lower().strip()
        clear_output()
    if ask == 'n':
        print('Okay, maybe next time!')
    if ask == 'y':
        action = input("What would you like to do? add/remove/show/clear/quit ").lower().strip()
        clear_output()
        while action not in ('add', 'remove', 'show', 'clear', 'quit'):
                action = input("Invalid response. Please enter add/remove/show/clear/quit ").lower().strip()
                clear_output()
        while action in ('add', 'remove', 'show', 'clear', 'quit'):
            if action == 'add':
                item = input("What would you like to put in your cart? ").lower().strip()
                clear_output()
                if item in shopping_cart:
                    quantity = input("How many of this item would you like to add? ").strip()
                    clear_output()
                    if quantity.isdigit():
                        quantity = int(quantity)
                    else:
                        while True:
                            quantity = int(input("Invalid response. Please enter a number.").strip())
                            clear_output()
                    shopping_cart[item][1] += quantity
                    total += shopping_cart[item][0] * quantity
                else:
                    price = float(input(f"How much does this item cost? ").strip('$ '))
                    clear_output()
                    quantity = input("How many of this item would you like to add? ").strip()
                    if quantity.isdigit():
                        quantity = int(quantity)
                    else:
                        while True:
                            quantity = int(input("Invalid response. Please enter a number.").strip())
                            clear_output()
                    if quantity > 1 and item[-1] == 's':
                        print(f"I am adding {quantity} {item} to your cart. ")
                        shopping_cart[item] = [price, quantity] 
                        total += price * quantity
                    elif quantity > 1:
                        print(f"I am adding {quantity} {item}s to your cart. ")
                        shopping_cart[item] = [price, quantity]
                        total += price * quantity
                    elif quantity == 1:
                        print(f"I am adding 1 {item} to your cart. ")
                        shopping_cart[item] = [price, quantity]
                        total += price * quantity
                    else:
                        print(f"I am removing the {item} from your cart.")
                    clear_output()
            if action == 'remove':
                item = input('What item would you like to remove? ').lower().strip()
                clear_output()
                while item not in shopping_cart:
                    item = input('This item is not in your cart. Please enter another item.').lower().strip()
                    clear_output()
                if shopping_cart[item][1] == 1:
                    print(f'You only have 1 {item} in your cart, which I am removing now.')
                    total -= shopping_cart[item][0]
                    del shopping_cart[item]
                if shopping_cart[item][1] > 1:
                    quantity = int(input(f'You have {shopping_cart[item][1]} of this item. How many would you like to remove? ').strip())
                    if quantity > shopping_cart[item][1]:
                        quantity = int(input('You do not have that many items in your cart. Please enter another number. ').strip())                        
                    if 0 < quantity < shopping_cart[item][1]:
                        shopping_cart[item][1] = shopping_cart[item][1] - quantity
                        total -= quantity * shopping_cart[item][0]
                        print(f'I have removed {quantity} of this item. You now have {shopping_cart[item][1]} left in your cart.')
                    elif quantity == shopping_cart[item][1]:
                        if item[-1] == 's':
                            total -= quantity * shopping_cart[item][0]
                            del shopping_cart[item]
                            print(f'I have removed all {item} from your cart.')
                        else:
                            total -= quantity * shopping_cart[item][0]
                            del shopping_cart[item]
                            print(f'I have removed all {item}s from your cart.')
                    elif quantity == 0:    
                        print("I will not remove any items from your cart.")
                    else:
                        quantity = int(input('Invalid response. Please enter another number. ').strip())
            if action == 'show':
                if shopping_cart == {}:
                    print('You have nothing in your cart!')
                else:
                    print(f"{name}'s cart: ")
                    for item in shopping_cart:
                        if item[-1] == 's':                   
                            print(f"{shopping_cart[item][1]} {item} for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")
                        else:
                            print(f"{shopping_cart[item][1]} {item}s for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")                                       
                    print('------------------------')
                    print(f"Total: ${total:.2f}")
            if action == 'clear':
                question = input("Are you sure you want to empty your cart? y/n ").lower().strip()
                while question not in ('y', 'n'):
                    question = input("Invalid response. Please enter y or n. ").lower().strip()
                    clear_output()
                if question == 'y':
                    shopping_cart.clear()
                    total = 0
                if question == 'n':
                    print('Thank goodness I asked!')
            if action == 'quit':
                print(f"{name}'s receipt: ")
                for item in shopping_cart:
                    if item[-1] == 's':                   
                        print(f"{shopping_cart[item][1]} {item} for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}") 
                    else:
                        print(f"{shopping_cart[item][1]} {item}s for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")                                        
                print('------------------------')
                print(f"Total: ${total:.2f}")
                break
            action = input('What else would you like to do? add/remove/show/clear/quit ').lower().strip()
            clear_output()
            while action not in ('add', 'remove', 'show', 'clear', 'quit'):
                action = input("Invalid response. Please enter add/remove/show/clear/quit ").lower().strip()
                clear_output()
            

    
go_shopping()

```
