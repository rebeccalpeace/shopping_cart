```python
from IPython.display import clear_output

# 1) takes in an input
# 2) stores user input into a dictionary or list
# 3) the user can add or delete items
# 4) the user can see current shopping list
# 5) the program loops until the user 'quits'
# 6) upon quitting the program, prints out a receipt of the items with total and quantity

def go_shopping():
    name = input('What is your name? ').title()
    ask = input(f"Hi {name}! I'm hungry. I'm gonna go shopping for some food!\nWould you like to join me? y/n ").lower().strip()
    clear_output()
    while ask not in ('y', 'n'):
        ask = input("Invalid response. Please enter y or n. ").lower().strip()
    if ask == 'n':
        print('Okay, maybe next time!')
    if ask == 'y':
        action = input("What would you like to do? add/remove/show/clear/quit ").lower().strip()
        clear_output()
        while action not in ('add', 'remove', 'show', 'clear', 'quit'):
                action = input("Invalid response. Please enter add/remove/show/clear/quit ").lower().strip()
        while action in ('add', 'remove', 'show', 'clear', 'quit'):
            if action == 'add':
                print(add_to_cart())   # helper function to add items to cart
            if action == 'remove':
                print(remove_from_cart())   # helper function to remove items from cart
            if action == 'show':
                # displays quantity, items, and price with total at the bottom
                if shopping_cart == {}:
                    print('You have nothing in your cart!')
                else:
                    total = show_total()
                    print(f"{name}'s cart: ")
                    for item in shopping_cart:
                        # for if item already ends in an 's'
                        if item[-1] == 's' or shopping_cart[item][1] == 1:                   
                            print(f"{shopping_cart[item][1]} {item} for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")
                        # for if item does not end in an 's'
                        else:
                            print(f"{shopping_cart[item][1]} {item}s for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")                                       
                    print('------------------------')
                    print(f"Total: ${total:.2f}")
            if action == 'clear':
                # this empties the shopping_cart dict completely and resets total to 0
                question = input("Are you sure you want to empty your cart? y/n ").lower().strip()
                while question not in ('y', 'n'):
                    question = input("Invalid response. Please enter y or n. ").lower().strip()
                if question == 'y':
                    shopping_cart.clear()
                    total = 0
                if question == 'n':
                    print('Thank goodness I asked!')
            if action == 'quit':
                # this prints out a list of quantity, items, and price with total at the bottom and then breaks to quit the program
                total = show_total()
                print(f"{name}'s receipt: ")
                for item in shopping_cart:
                    if item[-1] == 's' or shopping_cart[item][1] == 1:                   
                        print(f"{shopping_cart[item][1]} {item} for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}") 
                    else:
                        print(f"{shopping_cart[item][1]} {item}s for ${shopping_cart[item][0] * shopping_cart[item][1]:.2f}")                                        
                print('------------------------')
                print(f"Total: ${total:.2f}")
                break
            # this question will take you back to the while loop on line 15
            action = input('What else would you like to do? add/remove/show/clear/quit ').lower().strip()
            while action not in ('add', 'remove', 'show', 'clear', 'quit'):
                action = input("Invalid response. Please enter add/remove/show/clear/quit ").lower().strip()

    


```

```python
def add_to_cart():
    item = input("What would you like to put in your cart? ").lower().strip()
    # if item has already been added to dict and user wants to add more
    if item in shopping_cart:
        how_many = input("How many of this item would you like to add? ").strip()
        quantity = to_int(how_many)  # this func checks for if input is digit and will return an integer
        shopping_cart[item][1] += quantity
        if quantity > 1 and item[-1] == 's':
            return f"I am adding {quantity} {item} to your cart. "
        elif quantity > 1:
            return f"I am adding {quantity} {item}s to your cart. "
        elif quantity == 1:
            return f"I am adding 1 {item} to your cart. "
        else:
            return "I will not put any more in your cart."
    # if item is new and has not yet been added to dict
    else:
        how_much = input(f"How much does this item cost? ").strip('$ ')
        price = to_float(how_much)   # this func checks for if input is digit/float and will return float
        how_many = input("How many of this item would you like to add? ").strip()
        quantity = to_int(how_many)   # this func checks for if input is digit and will return an integer
        if quantity > 1 and item[-1] == 's':
            shopping_cart[item] = [price, quantity] 
            return f"I am adding {quantity} {item} to your cart."          
        elif quantity > 1:
            shopping_cart[item] = [price, quantity] 
            return f"I am adding {quantity} {item}s to your cart."
        elif quantity == 1:
            shopping_cart[item] = [price, quantity] 
            return f"I am adding 1 {item} to your cart."
        else:
            return f"I am removing the {item} from your cart."
```

```python
def remove_from_cart():
    item = input('What item would you like to remove? ').lower().strip()
    while item not in shopping_cart:
        item = input('This item is not in your cart. Please enter another item.').lower().strip()
    if shopping_cart[item][1] == 1:
        del shopping_cart[item]
        return f'You only have 1 {item} in your cart, which I am removing now.'
    if shopping_cart[item][1] > 1:
        quantity = to_int(input(f'You have {shopping_cart[item][1]} of this item. How many would you like to remove? ').strip())
        while quantity > shopping_cart[item][1]:
            quantity = to_int(input('You do not have that many items in your cart. Please enter another number. ').strip())                        
        if 0 < quantity < shopping_cart[item][1]:
            shopping_cart[item][1] = shopping_cart[item][1] - quantity
            return f'I have removed {quantity} of this item. You now have {shopping_cart[item][1]} left in your cart.'
        elif quantity == shopping_cart[item][1]:
            if item[-1] == 's':
                del shopping_cart[item]
                return f'I have removed all {item} from your cart.'
            else:
                del shopping_cart[item]
                return f'I have removed all {item}s from your cart.'
        else:    
            return "I will not remove any items from your cart."
```

```python
def to_float(string):
    price = string
    is_float = False
    while is_float == False:
        try:
            float(price)
            return float(price)
        except ValueError:
            price = input("Invalid response. Please enter a number.").strip('$ ')
```

```python
def to_int(string):
    if string.isdigit():
        return int(string)
    else:
        while string.isdigit() == False:
            string = input("Invalid response. Please enter a number.").strip()
        return int(string)
```

```python
def show_total():
    total = 0
    for item in shopping_cart:
        total += shopping_cart[item][0] * shopping_cart[item][1]
    return total
```
